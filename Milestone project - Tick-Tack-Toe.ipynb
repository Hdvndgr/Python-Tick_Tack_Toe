```python
from IPython.display import clear_output
params = {"Player 1":None, "Player 2":None, "isStarted":False, "Turn":0, "Winner":None}
tileList = list(9*' ')

def initGame():
    clear_output()
    ans =  str(input("To start the game; please, choose one of them 'X' or 'O'?  "))
    if ans.upper() == "X":
        params["Player 1"] = "X"
        params["Player 2"] = "O"
    elif ans.upper() == "O":
        params["Player 1"] = "O"
        params["Player 2"] = "X"
    else:
        print("You pressed another key.\n")
        initGame()
    startGame()
    

def startGame():
    draw(' ')
    makeMove()
    
def draw(element,tile=1):
    clear_output()
    if element == ' ':
        for it in range(len(tileList)):
            tileList[it] = ' '
    tileList[tile-1] = element
    index = 9
    for num in range(0,3):
        print('{0:^5}|{1:^5}|{2:^5}'.format(' ', ' ', ' '))
        print('{0:^5}|{1:^5}|{2:^5}'.format(tileList[index-3], tileList[index-2], tileList[index-1]))
        index = index-3
        if num == 2:
            print('{0:^5}|{1:^5}|{2:^5}'.format(' ', ' ', ' '))
        else:
            print('{0:^5}|{1:^5}|{2:^5}'.format('_____', '_____', '_____'))

def makeMove():
    if params["Turn"] == 0:
        move = int(input("\nPlayer 1's Turn: For making move please enter a number 1-9. "))
        draw(params["Player 1"],move)
        match = checkMatch()
        params["Turn"] = 1
        if match:
            endGame()
        else:
            if isFinished():
                endGame(True)
            else:
                makeMove()
    else:
        move = int(input("\nPlayer 2's Turn: For making move please enter a number 1-9. "))
        draw(params["Player 2"],move)
        match = checkMatch()
        params["Turn"] = 0
        if match:
            endGame()
        else:
            if isFinished():
                endGame(True)
            else:
                makeMove()
        
def checkMatch():
    checkList = [(0,1,2), (3,4,5), (6,7,8), (0,3,6), (1,4,7), (2,5,8), (0,4,8), (2,4,6)]
    for item1,item2,item3 in checkList:
        if tileList[item1] == tileList[item2] == tileList[item3] and tileList[item3] != ' ':
            params["Winner"] = params["Turn"]
            return True
    else:
        params["Winner"] = None 
        return False
def isFinished():
    result = None
    for value in tileList:
        if value == ' ':
            result = False
            break
        else:
            result = True
    return result
def endGame(apple=False):
    '''
    If apple == True it means the match is over but no one win.
    '''
    if apple:
        print("Nobody won! The match ended in a draw.\n")
    else:
        if params["Winner"]==0:
            print("Player 1 won the game.\n")
        elif params["Winner"]==1:
            print("Player 2 won the game.\n")
        else: 
            print("There is a problem. The game is restarting...")
    train = str(input("Do you wanna play again? Y or N   "))
    if train.upper() == 'Y':
        initGame()
    else:
        print("That's the game!")
        
        
```
